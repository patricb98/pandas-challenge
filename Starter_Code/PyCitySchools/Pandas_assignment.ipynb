```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
combined_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
combined_data.head()
```

```python
#find total number of schools list of schools
total_schools = len(combined_data["school_name"].unique())
total_schools 

```

```python
#calc total number of students
total_students = len(combined_data["student_name"])
total_students
```

```python
#calc total budget
unique_budget = combined_data["budget"].unique()
Total_budget = unique_budget.sum()
Total_budget
```

```python
#calc average math score
av_math = combined_data["maths_score"].mean()
av_math
```

```python
#calc average reading score
av_read = combined_data["reading_score"].mean()
av_read
```

```python
# % passing maths (the percentage of students who passed maths)
pass_math= combined_data[(combined_data["maths_score"]>= 50)].count()["Student ID"]
passing_math = (pass_math / total_students) * 100
passing_math

```

```python
#% passing reading (the percentage of students who passed reading)
pass_read= combined_data[(combined_data["reading_score"]>= 50)].count()["Student ID"]
passing_read = (pass_read / total_students) * 100
passing_read
```

```python
#% overall passing (the percentage of students who passed maths AND reading)
pass_both = combined_data[(combined_data["maths_score"]>= 50) & 
            (combined_data["reading_score"]>= 50)].count()["Student ID"]
pass_both_per = (pass_both/total_students) * 100
pass_both_per
```

```python
#formating DF to display table with results 
summary_df = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students], 
                           "Total Budget": [Total_budget], "Average Maths Score": [av_math], 
                           "Average Reading Score": [av_read], "% Passing Math": [passing_math],
                           "% Passing Reading": [passing_read], "% Overall Passing": [pass_both_per]})



summary_df

```

```python
# Group by school name
grouped_schools_df = combined_data.groupby("school_name")
```

```python
# School type
school_type = grouped_schools_df["type"].first()
school_type
```

```python
# Total students per school
total_student_per_school = combined_data["school_name"].value_counts()
total_student_per_school
```

```python
# Total School Budget
total_school_budget = grouped_schools_df["budget"].mean()
total_school_budget
```

```python
# Budget per student
budget_per_student = total_school_budget / total_student_per_school
budget_per_student
```

```python
# Average Maths Score
av_maths_score = grouped_schools_df["maths_score"].mean()
av_maths_score
```

```python
# Average reading score
av_reading_score = grouped_schools_df["reading_score"].mean()
av_reading_score
```

```python
# Passing maths, reading, and both
passing_math = combined_data[combined_data["maths_score"] >= 50].groupby("school_name").size()
passing_reading = combined_data[combined_data["reading_score"] >= 50].groupby("school_name").size()
passing_both = combined_data[(combined_data["maths_score"] >= 50) & (combined_data["reading_score"] >= 50)].groupby("school_name").size()
passing_both
```

```python
# % passing maths for each school
per_school_passing_math = (passing_math / total_student_per_school) * 100
per_school_passing_math
```

```python
# % passing reading for each school
per_school_passing_reading = (passing_reading / total_student_per_school) * 100
per_school_passing_reading
```

```python
# % overall passing per school
per_school_passing_overall = (passing_both / total_student_per_school) * 100
per_school_passing_overall
```

```python
# Create the DataFrame
school_summary_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_student_per_school,
    "Total School Budget": total_school_budget,
    "Per Student Budget": budget_per_student,
    "Average Maths Score": av_maths_score,
    "Average Reading Score": av_reading_score,
    "% Passing Maths": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": per_school_passing_overall
})
```

```python
#Create bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_summary_df['Spending Ranges (Per Student)'] = pd.cut(budget_per_student, spending_bins, labels=labels)
school_summary_df
```

```python
#calculate mean scores 
spending_math_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
spending_reading_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean()
spending_passing_reading = school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
#display

spending_summary = pd.DataFrame({
    "Average Maths Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Maths": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "overall_passing_spending": overall_passing_spending
})

spending_summary
```

```python
#Use the following code to bin the per_school_summary.

#size_bins = [0, 1000, 2000, 5000]
#labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
#Use pd.cut on the "Total Students" column of the per_school_summary DataFrame.

#Create a DataFrame called size_summary that breaks down school performance based on school size (small, medium, or large).
```

```python
#Create bins
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_summary_df['School Size'] = pd.cut(school_summary_df["Total Students"], size_bins, labels=labels)
school_summary_df
```

```python
#calculate mean scores 
size_math_scores = school_summary_df.groupby(["School Size"])["Average Maths Score"].mean()
size_reading_scores = school_summary_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_summary_df.groupby(["School Size"])["% Passing Maths"].mean()
size_passing_reading = school_summary_df.groupby(["School Size"])["% Passing Reading"].mean()
overall_passing_size = school_summary_df.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
#display

size_summary = pd.DataFrame({
    "Average Maths Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Maths": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "overall_passing_spending": overall_passing_size
})

size_summary
```
